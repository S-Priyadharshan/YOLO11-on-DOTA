```python
from ultralytics import YOLO

model = YOLO("yolo11n-obb.pt")

results = model.train(data="DOTAv1.5.yaml",epochs=10,imgsz=640)
```

```python
trained = YOLO("runs/obb/train/weights/best.pt")
```

```python
predictions = trained.predict(source="test.webp",task="obb",imgsz=640,save=True)
```
